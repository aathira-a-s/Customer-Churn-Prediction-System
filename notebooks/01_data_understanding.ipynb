# Customer Churn Prediction System

## Phase 2: Data Collection and Understanding

This notebook performs initial data understanding of the Telco Customer Churn dataset, including dataset structure analysis, feature understanding, target variable identification, missing value detection, and data type analysis.


## Activity 1 - Load Dataset
Load the Telco Customer Churn dataset into a pandas DataFrame and preview the first 5 records to confirm it loaded correctly.

```python
import pandas as pd

# Load dataset
df = pd.read_csv('../data/telco_customer_churn.csv')

# Preview first 5 records
df.head()
```

**Observation:** The dataset loaded successfully. The first 5 rows confirm that column names, values, and formatting are as expected for a telecom customer dataset.

## Activity 2 - Review Dataset Structure
Inspect the shape, column names, and general information of the dataset to understand its size and composition.

```python
print("=== Dataset Shape ===")
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")

print("\n=== Column Names ===")
print(df.columns.tolist())

print("\n=== Dataset Information ===")
df.info()
```

**Observation:** The dataset has 7,043 rows and 21 columns. All column names are meaningful business terms. The target variable is `Churn`. The `customerID` column is an identifier and will be dropped before model training.

## Activity 3 - Understand Feature Definitions
Generate numerical and categorical feature summaries to understand value distributions, ranges, and unique value counts for each feature.

```python
print("=== Numerical Feature Summary ===")
display(df.describe())

print("\n=== Categorical Feature Summary ===")
display(df.describe(include='object'))
```

**Observation:** There are 3 numerical features - `SeniorCitizen`, `tenure`, and `MonthlyCharges`. Tenure ranges from 0 to 72 months. MonthlyCharges ranges from `$18.25` to `$118.75`. There are 18 categorical (object) features. Note: `TotalCharges` is missing from the numerical summary because it is incorrectly stored as a string - this will be fixed during preprocessing.

## Activity 4 - Identify Target Variable
Identify the target variable `Churn`, examine its unique values, class distribution, and class percentages to understand the prediction task.

```python
print("=== Target Variable ===")
print("Churn")

print("\n=== Unique Values ===")
print(df['Churn'].unique())

print("\n=== Churn Class Counts ===")
print(df['Churn'].value_counts())

print("\n=== Churn Class Percentages ===")
print((df['Churn'].value_counts(normalize=True) * 100).round(2))
```

**Observation:** The target variable `Churn` has exactly 2 values - `Yes` (churned) and `No` (retained), confirming this is a binary classification problem. The dataset is moderately imbalanced: 73.46% No and 26.54% Yes. This means Recall will be prioritised over Accuracy as the primary evaluation metric during model training.

## Activity 5 - Detect Missing Values
Check for both formal null (NaN) values and hidden blank whitespace entries, which pandas does not detect as null by default.

```python
print("=== Missing Values Per Column ===")
display(df.isnull().sum())

print("\n=== Total Missing Values ===")
print(df.isnull().sum().sum())

print("\n=== Blank Values in Dataset ===")
display((df == " ").sum())

print("\n=== Blank Values in TotalCharges ===")
print((df["TotalCharges"] == " ").sum())
```

**Observation:** No formal null (NaN) values exist in any column. However, `TotalCharges` contains 11 blank whitespace entries (' ') that are not detected by `isnull()`. These must be handled separately during preprocessing using `pd.to_numeric(df['TotalCharges'], errors='coerce')`, which converts blanks to NaN, followed by median imputation.

## Activity 6 - Analyze Data Types
Inspect the data type of each column, count how many columns belong to each type, and separate features into numerical and categorical groups.

```python
print("=== Data Types ===")
display(df.dtypes)

print("\n=== Data Type Counts ===")
display(df.dtypes.value_counts())

numerical_features = df.select_dtypes(
    include=['int64', 'float64']
).columns

categorical_features = df.select_dtypes(
    include=['object']
).columns

print("\n=== Numerical Features ===")
print(list(numerical_features))

print("\n=== Categorical Features ===")
print(list(categorical_features))
```

**Observation:** pandas detects 3 numerical features (SeniorCitizen, tenure, MonthlyCharges) and 18 object features. `TotalCharges` incorrectly appears in the categorical list due to its data type error. After fixing it with `pd.to_numeric()`, it will move to the numerical group, bringing the correct numerical count to 4.

## Feature Summary Table
A combined summary showing the data type, missing value count, and number of unique values for all 21 features.

```python
feature_summary = pd.DataFrame({
    'Data Type': df.dtypes,
    'Missing Values': df.isnull().sum(),
    'Unique Values': df.nunique()
})

feature_summary
```

**Observation:** The feature summary confirms that only `TotalCharges` has a data quality issue (wrong data type + 11 blank entries). All other 20 columns are complete with no missing values. `customerID` has 7,043 unique values (one per row), confirming it is a row identifier that must be dropped before training.

## Key Findings

- The dataset contains 7,043 customer records and 21 features.
- `Churn` is the target variable and represents a binary classification problem (Yes / No).
- The dataset contains both numerical features (SeniorCitizen, tenure, MonthlyCharges) and categorical features (18 object columns).
- No formal null (NaN) values are present in any column.
- `TotalCharges` contains 11 blank whitespace entries and is incorrectly stored as object type - both issues must be fixed during preprocessing.
- The dataset is moderately class-imbalanced: 73.46% No Churn vs 26.54% Churn - Recall will be the primary evaluation metric.
- The dataset is suitable for customer churn prediction using machine learning techniques.
- Additional preprocessing and feature engineering will be performed before model development.
